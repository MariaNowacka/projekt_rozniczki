***Część Pierwsza.***

Lot rakiety batalistyczne wyróżnia się m.in.
- krótkim czasem trwania
- stosunkowo niedużą wysokością
- torem po parabolicznej krzywej balistycznej

oraz tym, że rakieta tylko przez część lotu jest zasilana silnikiem.

Jak zostało wspomniane w artykule [[5](#bibliografia)] na rakietę  działają następujace siły
- siła ciągu silnika rakiety
- siła oporu powietrza
- grawitacja
- siła dośrodkowa
- siła Coriolisa

W naszym uproszczonym modelu uznaliśmy, że na dosyć małej wysokości, przy prędkości o wiele mniejszej od prędkości kątowej Ziemi i krótkim czasie lotu, ruch rakiety zależy od siły Coriolisa w bardzo małym stopniu, i nie uwzględniamy tej siły w naszych obliczeniach. 

Rakietę traktujemy jako punkt materialny o dwóch współrzędnych (w pionie i poziomie), więc siła dośrodkowa również traci znaczenie.

W naszym modelu rakieta jest zależna od:
- siły ciągu rakiety
- oporu powietrza
- siły grawitacyjnej.

iły te opisaliśmy wzorami:
$F_T = T$
$F_op = -\frac{\rho_0 v^2C_DS}{2}$
$F_g = -mg$

Po przyrównaniu sumy tych sił do siły wypadkowej $F_w = ma$ uzyskujemy równanie:

$F_w = F_t + F_op + F_g$

co jest równoważne

$ ma = T -\frac{\rho_0 v^2C_DS}{2} -mg$

Dzieląc przez masę uzyskujemy równanie:

$ a = \frac{T -\frac{\rho_0 v^2C_DS}{2} -mg}{m}$

które można przedtawić jako poniższe równanie różniczkowe: 

$ \ddot{r} = \dot{v} =  \frac{ T - mg - 0.5\cdot \rho_0 v^2 C_d S}{m} $



Rozdzielając powyższe równanie na współrzędne $\ x, y\ $ otrzymujemy

$$ \ddot{r}_y = \dot{v}_y =  \frac{ T\sin(\theta) - mg - 0.5\cdot \rho_0 v_y^2 C_d S}{m} $$

$$ \ddot{r}_x = \dot{v}_x = \frac{T \cos(\theta) - 0.5\cdot \rho_0 v_x^2C_dS}{m}
$$
gdzie:

$
r_y \text{ - współrzędna pionowa położenia} \\
v_y \text{ - współrzędna pionowa prędkości} \\
r_x \text{ - współrzędna pozioma położenia} \\
v_x \text{ - współrzędna pozioma prędkości} \\
\theta \text{ - kąt odchylenia rakiety od pionu} \\
m \text{ - masa rakiety (zależna od czasu)} \\
g = 9.81 \frac{m}{s^2}\text{ - stała grawitacyjna } \\
S \text{ - powierzchnia przekroju poprzecznego rakiety} \\
\rho_0 = 1.22 \frac{kg}{m^3} \text{ - stała gęstości powietrza (wartość dla poziomu morza)} \\
C_d \text{ - współczynnik oporu powietrza} \\
T \text{ - siła ciągu silnika rakietowego} \\
$



**Część druga**

Do rozwiązania powyższych równań posłużymy się medodami numerycznymi - Eulera oraz Rungego-Kutty czwartego rzędu.

Metoda Eulera to bardzo podstawowa metoda służąca do rozwiązywania równań różniczkowych pierwszego rzędu (w naszym przypadku wystraczy zapisać równanie drugiego rzędu jako dwa równanie pierwszego rzędu), opierająca się na interpretacji geomertrycznej równania różniczkowego.

Pierwszym krokiem w tej metodzie jest zapisanie równania w postaci

$$ y' = f(x,y) $$

Drugim krokiem jest wyznaczenie początkowej wartości $y_0$ dla odpowiedniego $x_0$ oraz skoku czasowego $h$. Wtedy kolejne punkty na osi X można obliczyć z zależności:

$x_{n+1} = x_n + h$

a wartości $y_n$ :

$y_{n+1} = y_n + hf(x_n,y_n)$

Poniżej widać implementację tej metody w języku Python.

```python
import numpy as np, matplotlib.pyplot as plt
```

```python
# Metoda Eulera
def euler_method(f, y0, t, params):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(1, n):
        y[i] = y[i-1] + dt * f(y[i-1], t[i-1], params)
    return y
```

Metoda Rungego-Kutty rzędu czwartego daje lepsze rezultaty od metody Eulera, jest szybka i używa dosyć prostych wzorów.

Rozwiązując równanie drugiego rzędu zaczynamy od napisania układu równań:

$$
\left\{ {y'' = v' \ \ \ \atop v' =f(x,v)} \right.
$$

Wtedy rozwiązujemy 2 równania I rzędu, zamiast 1 równania II rzędu - jest to łatwiejsza metoda.

Podobnie jak w metodzie Eulera dyskretyzujemy zmienne, ustalając $x_0$, $y_0$ oraz krok czasowy $h$, a w każdym kolejnym, $n$-tym kroku iteracji obliczamy:


$ x_{n+1} = x_n +h \\
k_1 = h \cdot f(x_n, y_n) \\
k_2 = h \cdot f(x_n + 0.5 \cdot h, y_n + 0.5 \cdot k_1) \\
k_3 = h \cdot f(x_n + 0.5 \cdot h, y_n + 0.5 \cdot k_2) \\
k_4 = h \cdot f(x_n + h, y_n + k_3) $

Następnie liczymy:

$dy_n = \frac{k_1+2k_2+2k_3+k_4}{6}$

i ostatecznie mamy:

$y_{n+1} = y_n + dy_n$

Co można przedstawić poniższym kodem.

```python
# Metoda Rungego-Kutty (RK4)
def runge_kutta_method(f, y0,t, params):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(1, n):
        k1 = dt * f(y[i-1], t[i-1], params)
        k2 = dt * f(y[i-1] + 0.5*k1, t[i-1] + 0.5*dt, params)
        k3 = dt * f(y[i-1] + 0.5*k2, t[i-1] + 0.5*dt, params)
        k4 = dt * f(y[i-1] + k3, t[i-1] + dt, params)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
    return y
```

***Symulacja.***

```python
#  wersja 1

import numpy as np
import matplotlib.pyplot as plt

# Funkcja opisująca równanie ruchu rakiety balistycznej
def rocket_equation(y, t, Cd, A, rho, g, mass_function):
    theta = np.radians(90)-1.01*t
    v, theta, m, = y
    dvdt = (-m * g * np.sin(theta) - 0.5 * rho * v**2 * Cd * A) / m
    dthetadt = 0
    dmdt = -mass_function(t)  # Zmiana masy w czasie
    return np.array([dvdt, dthetadt, dmdt])

# Funkcja opisująca równanie drogi
def path_equation(y, t, Cd, A, rho, g):
    v, theta, m = y
    dsdt = v * np.cos(theta)
    return dsdt

# Metoda Eulera
def euler_method(rocket_eq, path_eq, y0, t, Cd, A, rho, g, mass_function):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    s = np.zeros(n)
    y[0] = y0
    for i in range(1, n):
        y[i] = y[i-1] + dt * rocket_eq(y[i-1], t[i-1], Cd, A, rho, g, mass_function)
        s[i] = s[i-1] + dt * path_eq(y[i-1], t[i-1], Cd, A, rho, g)
    return s, -y[:, 0]

# Metoda Rungego-Kutty (RK4)
def runge_kutta_method(rocket_eq, path_eq, y0, t, Cd, A, rho, g, mass_function):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    s = np.zeros(n)
    y[0] = y0
    for i in range(1, n):
        k1 = dt * rocket_eq(y[i-1], t[i-1], Cd, A, rho, g, mass_function)
        k2 = dt * rocket_eq(y[i-1] + 0.5*k1, t[i-1] + 0.5*dt, Cd, A, rho, g, mass_function)
        k3 = dt * rocket_eq(y[i-1] + 0.5*k2, t[i-1] + 0.5*dt, Cd, A, rho, g, mass_function)
        k4 = dt * rocket_eq(y[i-1] + k3, t[i-1] + dt, Cd, A, rho, g, mass_function)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
        s[i] = s[i-1] + dt * path_eq(y[i-1], t[i-1], Cd, A, rho, g)
    return s, -y[:, 0]

# Funkcja zmiany masy w czasie
def mass_function(t):
    if 2800 - 57 *t >=0:
        return 5900-(2800-57*t)
    else:
        return 3100

# Parametry rakiety
Cd = 0.5
A = 1.0
rho = 1.2
g = 9.81

# Warunki początkowe
initial_velocity = 450
initial_angle = np.radians(45)
initial_mass = 5900.0
initial_conditions = np.array([initial_velocity, initial_angle, initial_mass])

# Czas symulacji
t = np.linspace(0, 150, 1100)

# Rozwiązanie równania ruchu numerycznie - Metoda Eulera
solution_euler, velocity_euler = euler_method(
    rocket_equation, path_equation, initial_conditions, t, Cd, A, rho, g, mass_function
)

# Rozwiązanie równania ruchu numerycznie - Metoda Rungego-Kutty (RK4)
solution_rk4, velocity_rk4 = runge_kutta_method(
    rocket_equation, path_equation, initial_conditions, t, Cd, A, rho, g, mass_function
)

# Wykres prędkości w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, velocity_euler, label='Euler')
plt.plot(t, velocity_rk4, label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prędkość (m/s)')
plt.title('Ruch rakiety balistycznej - Prędkość w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres drogi w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler, label='Euler')
plt.plot(t, solution_rk4, label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Dystans (m)')
plt.title('Ruch rakiety balistycznej - Droga w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# wersja 2
import numpy as np, math
import matplotlib.pyplot as plt

def fun_T(i):
    if 2800 - 57*i<0:
        return 0
    else:
        return 2700

# Funkcja opisująca równanie ruchu rakiety balistycznej
def rocket_equation(y, t, Cd, A, rho, g, mass_function):
    theta = np.radians(90)
    T= fun_T(t)
    v, theta, m = y
    dvdt = -(T*math.sin(theta)-m * g * np.sin(theta) + 0.5 * rho * v**2 * Cd * A) / m
    dthetadt = -0.1*theta
    dmdt = -mass_function(t)  # Zmiana masy w czasie
    return np.array([dvdt, dthetadt, dmdt])

# Funkcja opisująca równanie drogi
def path_equation(y, t, Cd, A, rho, g):
    v, theta, m = y
    dsdt = v * np.cos(theta)
    return dsdt

# Metoda Eulera
def euler_method(rocket_eq, path_eq, y0, t, Cd, A, rho, g, mass_function):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    s = np.zeros(n)
    y[0] = y0
    for i in range(1, n):
        y[i] = y[i-1] + dt * rocket_eq(y[i-1], t[i-1], Cd, A, rho, g, mass_function)
        s[i] = s[i-1] + dt * path_eq(y[i-1], t[i-1], Cd, A, rho, g)
    return s, y[:, 0]

# Metoda Rungego-Kutty (RK4)
def runge_kutta_method(rocket_eq, path_eq, y0, t, Cd, A, rho, g, mass_function):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    s = np.zeros(n)
    y[0] = y0
    for i in range(1, n):
        k1 = dt * rocket_eq(y[i-1], t[i-1], Cd, A, rho, g, mass_function)
        k2 = dt * rocket_eq(y[i-1] + 0.5*k1, t[i-1] + 0.5*dt, Cd, A, rho, g, mass_function)
        k3 = dt * rocket_eq(y[i-1] + 0.5*k2, t[i-1] + 0.5*dt, Cd, A, rho, g, mass_function)
        k4 = dt * rocket_eq(y[i-1] + k3, t[i-1] + dt, Cd, A, rho, g, mass_function)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
        s[i] = s[i-1] + dt * path_eq(y[i-1], t[i-1], Cd, A, rho, g)
    return s, y[:, 0]

# Funkcja zmiany masy w czasie
def mass_function(t):
    if 2800 - 57 *t >=0:
        return 5900-(2800-57*t)
    else:
        return 3100

# Parametry rakiety
Cd = 0.5
A = 1.0 * math.pi
rho = 1.2
g = 9.81

# Warunki początkowe
initial_velocity = 400
initial_angle = np.radians(89)
initial_mass = 5900.0
initial_conditions = np.array([initial_velocity, initial_angle, initial_mass])

# Czas symulacji
t = np.linspace(0, 300, 3000)

# Rozwiązanie równania ruchu numerycznie - Metoda Eulera
solution_euler, velocity_euler = euler_method(
    rocket_equation, path_equation, initial_conditions, t, Cd, A, rho, g, mass_function
)

# Rozwiązanie równania ruchu numerycznie - Metoda Rungego-Kutty (RK4)
solution_rk4, velocity_rk4 = runge_kutta_method(
    rocket_equation, path_equation, initial_conditions, t, Cd, A, rho, g, mass_function
)

# Wykres prędkości w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, velocity_euler, label='Euler')
plt.plot(t, velocity_rk4, label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prędkość (m/s)')
plt.title('Ruch rakiety balistycznej - Prędkość w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres drogi w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler, label='Euler')
plt.plot(t, solution_rk4, label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Dystans (m)')
plt.title('Ruch rakiety balistycznej - Droga w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# wersja 3

import numpy as np
import matplotlib.pyplot as plt

# Parametry rakiety
m0 = 5900  # początkowa masa rakiety [kg]
mdot = 57  # tempo spalania paliwa [kg/s]
g = 9.81  # przyspieszenie ziemskie [m/s^2]
Thrust = 160000  # zwiększony ciąg silnika [N]
burnout_mass = 2800  # masa rakiety po wyczerpaniu paliwa [kg]
rho = 1.2  # Gęstość powietrza [kg/m^3]
Cd = 1.1  # Zmniejszony współczynnik oporu aerodynamicznego
A = 3.5  # Powierzchnia przekroju poprzecznego rakiety [m^2]

# Warunki początkowe
v0 = 10  # początkowa prędkość [m/s]
y0 = 0  # początkowa wysokość [m]
theta0 = 90  # początkowy kąt [stopnie] (rakieta zwrócona do góry)
t0 = 0  # początkowy czas [s]


# Funkcje opisujące równania ruchu
def mass(t):
    return m0 - mdot * t if m0 - mdot * t >= burnout_mass else burnout_mass


def rocket_eq(t, state):
    m = mass(t)
    # Zmiana kąta po 20 sekundach
    if t >= 25:
        dthetadt = np.radians(-1.2)  # Prędkość zmiany kąta
    else:
        dthetadt = 0  # Kąt nie zmienia się

    dvdt = ((-m * g - 0.5 * rho * state[0] * np.abs(state[0]) * Cd * A + Thrust) / m) if m > burnout_mass else -g
    dydt = state[0]

    return np.array([dvdt, dydt, dthetadt])


# Metoda Eulera do rozwiązania równań różniczkowych
def euler_method(h, T):
    num_steps = int(T / h)
    times = np.linspace(t0, T, num_steps + 1)
    states = np.zeros((num_steps + 1, 3))
    states[0, :] = [v0, y0, np.radians(theta0)]  # Zamiana stopni na radiany

    for i in range(num_steps):
        t = times[i]
        y = states[i, :]
        states[i + 1, :] = y + h * rocket_eq(t, y)

        # Sprawdzenie warunku przerwania
        if states[i + 1, 1] <= 0:
            break

    return times[:i + 2], states[:i + 2]


# Parametry symulacji
h = 0.1  # krok czasowy [s]
T = 120 # czas trwania symulacji [s]

# Rozwiązanie równań ruchu
times, states = euler_method(h, T)

# Wykresy
plt.figure(figsize=(12, 7))
plt.subplot(3, 1, 1)
plt.plot(times, states[:, 0])
plt.title('Prędkość rakiety w funkcji czasu')
plt.xlabel('Czas [s]')
plt.ylabel('Prędkość [m/s]')

plt.subplot(3, 1, 2)
plt.plot(times, states[:, 1])
plt.title('Wysokość rakiety w funkcji czasu')
plt.xlabel('Czas [s]')
plt.ylabel('Wysokość [m]')

plt.subplot(3, 1, 3)
plt.plot(times, np.degrees(states[:, 2]))  # Przeliczenie radianów na stopnie
plt.title('Kąt rakiety w funkcji czasu')
plt.xlabel('Czas [s]')
plt.ylabel('Kąt [stopnie]')

plt.tight_layout()
plt.show()
```

```python
# upośledzona wersja 4

import numpy as np, math
import matplotlib.pyplot as plt

m_loss = 57.8
fuel_mass = 2800
rocket_mass = 5900.0
max_thrust = 1000
no_fuel = fuel_mass/m_loss
S = 1.5**2*math.pi

params = {'Cd' : 0.5, 'S' : 1.5**2*math.pi, 'rho' : 1.2, 'g' : 9.81}

# Funkcja zmiany masy w czasie
def mass_function(t):
    if fuel_mass - m_loss *t > 0:
        return rocket_mass -(m_loss * t)
    else:
        return rocket_mass - fuel_mass

def fun_T(t):
    if t < no_fuel:
        return max_thrust
    else:
        return 0

# Funkcja opisująca równanie ruchu rakiety balistycznej
def rocket_equation(var, t, params):
    vx, vy, x, y, m = var
    g, rho, S, Cd = params['g'], params['rho'], params['S'], params['Cd']
    
    T = fun_T(t)
    theta = np.arctan2(vy,vx)
    v = np.sqrt(vx**2 + vy**2)

    dvxdt = (T * np.cos(theta) - 0.5 * rho * v**2 * Cd * S) / m
    dvydt = (T * np.sin(theta) - m * g * np.cos(theta) - 0.5 * rho * v**2 * Cd * S) / m
    dxdt = vx
    dydt = vy
    dmdt = - m_loss if t < no_fuel else 0

    return np.array([dvxdt, dvydt, dxdt, dydt, dmdt])

# Metoda Eulera
def euler_method(f, y0, t, params):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(1, n):
        y[i] = y[i-1] + dt * f(y[i-1], t[i-1], params)
    return y

# Metoda Rungego-Kutty (RK4)
def runge_kutta_method(f, y0,t, params):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(1, n):
        k1 = dt * f(y[i-1], t[i-1], params)
        k2 = dt * f(y[i-1] + 0.5*k1, t[i-1] + 0.5*dt, params)
        k3 = dt * f(y[i-1] + 0.5*k2, t[i-1] + 0.5*dt, params)
        k4 = dt * f(y[i-1] + k3, t[i-1] + dt, params)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
    return y

# Warunki początkowe
theta = 45
initial_velocity = 450
vx0 = initial_velocity * np.cos(np.radians(theta))
vy0 = initial_velocity * np.sin(np.radians(theta))
x0, y0 = 0, 0
initial_conditions = np.array([vx0, vy0, x0, y0, rocket_mass])

# Czas symulacji
t_k = 85
t = np.linspace(0, t_k, int(t_k / 0.5 + 1))


# Rozwiązanie równania ruchu numerycznie - Metoda Eulera
solution_euler = euler_method(
    rocket_equation, initial_conditions, t, params)

# Rozwiązanie równania ruchu numerycznie - Metoda Rungego-Kutty (RK4)
solution_rk4 = runge_kutta_method(
    rocket_equation, initial_conditions, t, params)

# Wykres przemieszczenia
plt.figure(figsize=(10, 6))

plt.plot(solution_euler[:,2], solution_euler[:,3], label='Euler')
plt.plot(solution_rk4[:,2], solution_rk4[:,3], label='RK4')
plt.xlabel('Dystans')
plt.ylabel('Wysokość')
plt.title('Ruch rakiety balistycznej - przemieszczenie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres prędkości x w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,0], label='Euler')
plt.plot(t, solution_rk4[:,0], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prędkość x (m/s)')
plt.title('Ruch rakiety balistycznej - Prędkość x w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres prędkości y w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,1], label='Euler')
plt.plot(t, solution_rk4[:,1], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prędkość y (m/s)')
plt.title('Ruch rakiety balistycznej - Prędkość y w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres prędkości w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, np.sqrt(solution_euler[:,0]**2 + solution_euler[:,1]**2), label='Euler')
plt.plot(t, np.sqrt(solution_rk4[:,0]**2 + solution_rk4[:,1]**2), label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prędkość (m/s)')
plt.title('Ruch rakiety balistycznej - Prędkość w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres drogi w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,2], label='Euler')
plt.plot(t, solution_rk4[:,2], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Dystans x (m)')
plt.title('Ruch rakiety balistycznej - Droga w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# wykres wysokości w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,3], label='Euler')
plt.plot(t, solution_rk4[:,3], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Wysokość y (m)')
plt.title('Ruch rakiety balistycznej - Wysokość w czasie')
plt.legend()
plt.grid(True)
plt.show()

# wykres masy
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,4], label='Euler')
plt.plot(t, solution_rk4[:,4], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('masa')
plt.title('ubytek masy')
plt.legend()
plt.grid(True)
plt.show()
```

W artykule [[1](#bibliografia)] przedstawiono model matematyczny lotu rakiety balistycznej w centralnym polu grawitacyjnym, opracowany z wykorzystaniem notacji macierzowej.
Model uwzględnia trzy charakterystyczne dla rakiet balistycznych fazy lotu: 
- startową (ang. boost phase), 
- środkową (ang. midcourse phase) i 
- terminalną (ang. terminal phase) 

oraz wpływ oddziaływania atmosfery i sił pozornych na ruch 
rakiet w przestrzeni. (siła odśrodkowa i Coriolisa)

Obliczenia trajektorii lotu rakiety balistycznej prowadzone są w kartezjańskim układzie 
współrzędnych, przedstawionym na rys. 1. Początek tego układu pokrywa się ze środkiem 
Ziemi, oś z skierowana jest do góry i przechodzi przez biegun północny, zaś osie x i y  leżą w płaszczyźnie równika, przy czym x przechodzi przez południk 0° (Greenwich). Przyjęty układ odniesienia jest układem nieinercjalnym, obraca się 
z prędkością kątową $\omega$ względem osi z, co implikuje konieczność uwzględnienia w rozważaniach sił: odśrodkowej i Coriolisa. Na potrzeby obliczeń zakłada się ponadto, że 
powierzchnia planety jest idealną sferą, zaś Ziemia charakteryzuje się następującymi 
parametrami fizycznymi: promieniem $R = 6,37·10^6 \ m$, masą $M = 5,97·10^{24}\ kg$ i prędkością obrotową $ω = 7,29·10^{-5} \frac{rad}{s} $. Wartość stałej grawitacyjnej wynosi $G = 6,67·10^{-11} \frac{m^3}{kg \cdot s}$.

W  modelu przedstawionym w artykule [[1](#bibliografia)] traktujemy rakietę jako punkt materialny o 3 stopniach swobody (ignorujemy stopnie swobody związane z obrotai w przestzreni). Wtedy przyspieszenie można przedstawić jako sumę:
$$
    \ddot{r} = \dot{v} =  a_T + a_C + a_D + a_A 
$$,
gdzie : 

$
r -\text{wektor położenia rakiety balistycznej} \\
v - \text{wektor prędkości rakiety balistycznej} \\
a_T - \text{wektor przyspieszenia związany z ciągiem silnika rakietowego} \\
a_C - \text{wektor przyspieszenia sterującego} \\
a_D - \text{wektor przyspieszenia związany z oporem atmosfery} \\
a_G - \text{wektor przyspieszenia grawitacyjnego} \\
a_A - \text{wektor przyspieszenia pozornego (odśrodkowego i Coriolisa)} 
$ 

Przyspieszenie związane z oporem powietrza (poniżej 100 km od powierzchni Ziemi) można opisać wzorem:

$$ a_D = - \frac{\rho_0 e^{-kh}c_D S}{2(m_0 - \int_0^{t_b} \dot{m_f} \operatorname{d}\!t)g}||v||v
$$,
gdzie:

$\rho_0 = 1,22 kg/m^3 - \text{gęstość atmosfery na wysokości 0 m n.p.m.,} \\
k = 0,14141\cdot 10^{-3} m^{-1} \text{ - współczynnik} \\
h - \text{wysokość lotu rakiety} \\
c_D - \text{współczynik oporu aerodynamicznego kadłuba} \\
S - \text{powierzchnia charakterystyczna} \\
m_0 - \text{masa stratowa rakiety} \\
\dot{m_f} - \text{wydatek masy paliwa} \\
t_b - \text{ czas pracy silnika rakietowego} \\
g - \text{przyspieszenie ziemskie}
$ 

Przyspieszenie grawitacyjne rakiety przedstawiono jako:

$$ a_G = - \frac{GMr}{||r||^3} $$

przyspieszenie pozorne :

$$ a_A = -2(\omega \times v)-\omega (\omega \times r) $$

przyspieszenie od siły ciągu:

$$ a_T = \frac{T}{(m_0 - \int_0^{t_b} \dot{m_f} \operatorname{d}\!t)} \frac{v}{||v||} $$

gdzie: $ T $ - ciąg silnika rakietowego,

a wektor przyspieszenia sterującego opisano jako:

$$ a_C = \frac{aT}{2}\left[1+\cos\left(\frac{\pi t}{t_b - t_v}\right)\right]\delta $$

gdzie:

$\delta$ - jednostkowy wektor sterujący,

$a$ - współczynnik korygujący,

$t_b$ - czas pracy silnika rakietowego,

$t_v$ - czas lotu pionowego,

co zapewnia płynną regulację kąta pochylenia $\gamma$ wektora prędkości w chwili wyłączenia napędu, co pokazano na rysunku 2 [[1](#bibliografia)].
