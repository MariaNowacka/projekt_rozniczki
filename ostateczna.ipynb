***Cz Pierwsza.***

Lot rakiety batalistyczne wyrnia si m.in.
- krtkim czasem trwania
- stosunkowo niedu wysokoci
- torem po parabolicznej krzywej balistycznej

oraz tym, e rakieta tylko przez cz lotu jest zasilana silnikiem.

Jak zostao wspomniane w artykule [[5](#bibliografia)] na rakiet  dziaaj nastpujace siy
- sia cigu silnika rakiety
- sia oporu powietrza
- grawitacja
- sia dorodkowa
- sia Coriolisa

W naszym uproszczonym modelu uznalimy, e na dosy maej wysokoci, przy prdkoci o wiele mniejszej od prdkoci ktowej Ziemi i krtkim czasie lotu, ruch rakiety zaley od siy Coriolisa w bardzo maym stopniu, i nie uwzgldniamy tej siy w naszych obliczeniach. 

Rakiet traktujemy jako punkt materialny o dwch wsprzdnych (w pionie i poziomie), wic sia dorodkowa rwnie traci znaczenie.

W naszym modelu rakieta jest zalena od:
- siy cigu rakiety
- oporu powietrza
- siy grawitacyjnej.

iy te opisalimy wzorami:
$F_T = T$
$F_op = -\frac{\rho_0 v^2C_DS}{2}$
$F_g = -mg$

Po przyrwnaniu sumy tych si do siy wypadkowej $F_w = ma$ uzyskujemy rwnanie:

$F_w = F_t + F_op + F_g$

co jest rwnowane

$ ma = T -\frac{\rho_0 v^2C_DS}{2} -mg$

Dzielc przez mas uzyskujemy rwnanie:

$ a = \frac{T -\frac{\rho_0 v^2C_DS}{2} -mg}{m}$

ktre mona przedtawi jako ponisze rwnanie rniczkowe: 

$ \ddot{r} = \dot{v} =  \frac{ T - mg - 0.5\cdot \rho_0 v^2 C_d S}{m} $



Rozdzielajc powysze rwnanie na wsprzdne $\ x, y\ $ otrzymujemy

$$ \ddot{r}_y = \dot{v}_y =  \frac{ T\sin(\theta) - mg - 0.5\cdot \rho_0 v_y^2 C_d S}{m} $$

$$ \ddot{r}_x = \dot{v}_x = \frac{T \cos(\theta) - 0.5\cdot \rho_0 v_x^2C_dS}{m}
$$
gdzie:

$
r_y \text{ - wsprzdna pionowa pooenia} \\
v_y \text{ - wsprzdna pionowa prdkoci} \\
r_x \text{ - wsprzdna pozioma pooenia} \\
v_x \text{ - wsprzdna pozioma prdkoci} \\
\theta \text{ - kt odchylenia rakiety od pionu} \\
m \text{ - masa rakiety (zalena od czasu)} \\
g = 9.81 \frac{m}{s^2}\text{ - staa grawitacyjna } \\
S \text{ - powierzchnia przekroju poprzecznego rakiety} \\
\rho_0 = 1.22 \frac{kg}{m^3} \text{ - staa gstoci powietrza (warto dla poziomu morza)} \\
C_d \text{ - wspczynnik oporu powietrza} \\
T \text{ - sia cigu silnika rakietowego} \\
$



**Cz druga**

Do rozwizania powyszych rwna posuymy si medodami numerycznymi - Eulera oraz Rungego-Kutty czwartego rzdu.

Metoda Eulera to bardzo podstawowa metoda suca do rozwizywania rwna rniczkowych pierwszego rzdu (w naszym przypadku wystraczy zapisa rwnanie drugiego rzdu jako dwa rwnanie pierwszego rzdu), opierajca si na interpretacji geomertrycznej rwnania rniczkowego.

Pierwszym krokiem w tej metodzie jest zapisanie rwnania w postaci

$$ y' = f(x,y) $$

Drugim krokiem jest wyznaczenie pocztkowej wartoci $y_0$ dla odpowiedniego $x_0$ oraz skoku czasowego $h$. Wtedy kolejne punkty na osi X mona obliczy z zalenoci:

$x_{n+1} = x_n + h$

a wartoci $y_n$ :

$y_{n+1} = y_n + hf(x_n,y_n)$

Poniej wida implementacj tej metody w jzyku Python.

```python
import numpy as np, matplotlib.pyplot as plt
```

```python
# Metoda Eulera
def euler_method(f, y0, t, params):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(1, n):
        y[i] = y[i-1] + dt * f(y[i-1], t[i-1], params)
    return y
```

Metoda Rungego-Kutty rzdu czwartego daje lepsze rezultaty od metody Eulera, jest szybka i uywa dosy prostych wzorw.

Rozwizujc rwnanie drugiego rzdu zaczynamy od napisania ukadu rwna:

$$
\left\{ {y'' = v' \ \ \ \atop v' =f(x,v)} \right.
$$

Wtedy rozwizujemy 2 rwnania I rzdu, zamiast 1 rwnania II rzdu - jest to atwiejsza metoda.

Podobnie jak w metodzie Eulera dyskretyzujemy zmienne, ustalajc $x_0$, $y_0$ oraz krok czasowy $h$, a w kadym kolejnym, $n$-tym kroku iteracji obliczamy:


$ x_{n+1} = x_n +h \\
k_1 = h \cdot f(x_n, y_n) \\
k_2 = h \cdot f(x_n + 0.5 \cdot h, y_n + 0.5 \cdot k_1) \\
k_3 = h \cdot f(x_n + 0.5 \cdot h, y_n + 0.5 \cdot k_2) \\
k_4 = h \cdot f(x_n + h, y_n + k_3) $

Nastpnie liczymy:

$dy_n = \frac{k_1+2k_2+2k_3+k_4}{6}$

i ostatecznie mamy:

$y_{n+1} = y_n + dy_n$

Co mona przedstawi poniszym kodem.

```python
# Metoda Rungego-Kutty (RK4)
def runge_kutta_method(f, y0,t, params):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(1, n):
        k1 = dt * f(y[i-1], t[i-1], params)
        k2 = dt * f(y[i-1] + 0.5*k1, t[i-1] + 0.5*dt, params)
        k3 = dt * f(y[i-1] + 0.5*k2, t[i-1] + 0.5*dt, params)
        k4 = dt * f(y[i-1] + k3, t[i-1] + dt, params)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
    return y
```

***Symulacja.***

```python
#  wersja 1

import numpy as np
import matplotlib.pyplot as plt

# Funkcja opisujca rwnanie ruchu rakiety balistycznej
def rocket_equation(y, t, Cd, A, rho, g, mass_function):
    theta = np.radians(90)-1.01*t
    v, theta, m, = y
    dvdt = (-m * g * np.sin(theta) - 0.5 * rho * v**2 * Cd * A) / m
    dthetadt = 0
    dmdt = -mass_function(t)  # Zmiana masy w czasie
    return np.array([dvdt, dthetadt, dmdt])

# Funkcja opisujca rwnanie drogi
def path_equation(y, t, Cd, A, rho, g):
    v, theta, m = y
    dsdt = v * np.cos(theta)
    return dsdt

# Metoda Eulera
def euler_method(rocket_eq, path_eq, y0, t, Cd, A, rho, g, mass_function):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    s = np.zeros(n)
    y[0] = y0
    for i in range(1, n):
        y[i] = y[i-1] + dt * rocket_eq(y[i-1], t[i-1], Cd, A, rho, g, mass_function)
        s[i] = s[i-1] + dt * path_eq(y[i-1], t[i-1], Cd, A, rho, g)
    return s, -y[:, 0]

# Metoda Rungego-Kutty (RK4)
def runge_kutta_method(rocket_eq, path_eq, y0, t, Cd, A, rho, g, mass_function):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    s = np.zeros(n)
    y[0] = y0
    for i in range(1, n):
        k1 = dt * rocket_eq(y[i-1], t[i-1], Cd, A, rho, g, mass_function)
        k2 = dt * rocket_eq(y[i-1] + 0.5*k1, t[i-1] + 0.5*dt, Cd, A, rho, g, mass_function)
        k3 = dt * rocket_eq(y[i-1] + 0.5*k2, t[i-1] + 0.5*dt, Cd, A, rho, g, mass_function)
        k4 = dt * rocket_eq(y[i-1] + k3, t[i-1] + dt, Cd, A, rho, g, mass_function)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
        s[i] = s[i-1] + dt * path_eq(y[i-1], t[i-1], Cd, A, rho, g)
    return s, -y[:, 0]

# Funkcja zmiany masy w czasie
def mass_function(t):
    if 2800 - 57 *t >=0:
        return 5900-(2800-57*t)
    else:
        return 3100

# Parametry rakiety
Cd = 0.5
A = 1.0
rho = 1.2
g = 9.81

# Warunki pocztkowe
initial_velocity = 450
initial_angle = np.radians(45)
initial_mass = 5900.0
initial_conditions = np.array([initial_velocity, initial_angle, initial_mass])

# Czas symulacji
t = np.linspace(0, 150, 1100)

# Rozwizanie rwnania ruchu numerycznie - Metoda Eulera
solution_euler, velocity_euler = euler_method(
    rocket_equation, path_equation, initial_conditions, t, Cd, A, rho, g, mass_function
)

# Rozwizanie rwnania ruchu numerycznie - Metoda Rungego-Kutty (RK4)
solution_rk4, velocity_rk4 = runge_kutta_method(
    rocket_equation, path_equation, initial_conditions, t, Cd, A, rho, g, mass_function
)

# Wykres prdkoci w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, velocity_euler, label='Euler')
plt.plot(t, velocity_rk4, label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prdko (m/s)')
plt.title('Ruch rakiety balistycznej - Prdko w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres drogi w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler, label='Euler')
plt.plot(t, solution_rk4, label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Dystans (m)')
plt.title('Ruch rakiety balistycznej - Droga w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# wersja 2
import numpy as np, math
import matplotlib.pyplot as plt

def fun_T(i):
    if 2800 - 57*i<0:
        return 0
    else:
        return 2700

# Funkcja opisujca rwnanie ruchu rakiety balistycznej
def rocket_equation(y, t, Cd, A, rho, g, mass_function):
    theta = np.radians(90)
    T= fun_T(t)
    v, theta, m = y
    dvdt = -(T*math.sin(theta)-m * g * np.sin(theta) + 0.5 * rho * v**2 * Cd * A) / m
    dthetadt = -0.1*theta
    dmdt = -mass_function(t)  # Zmiana masy w czasie
    return np.array([dvdt, dthetadt, dmdt])

# Funkcja opisujca rwnanie drogi
def path_equation(y, t, Cd, A, rho, g):
    v, theta, m = y
    dsdt = v * np.cos(theta)
    return dsdt

# Metoda Eulera
def euler_method(rocket_eq, path_eq, y0, t, Cd, A, rho, g, mass_function):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    s = np.zeros(n)
    y[0] = y0
    for i in range(1, n):
        y[i] = y[i-1] + dt * rocket_eq(y[i-1], t[i-1], Cd, A, rho, g, mass_function)
        s[i] = s[i-1] + dt * path_eq(y[i-1], t[i-1], Cd, A, rho, g)
    return s, y[:, 0]

# Metoda Rungego-Kutty (RK4)
def runge_kutta_method(rocket_eq, path_eq, y0, t, Cd, A, rho, g, mass_function):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    s = np.zeros(n)
    y[0] = y0
    for i in range(1, n):
        k1 = dt * rocket_eq(y[i-1], t[i-1], Cd, A, rho, g, mass_function)
        k2 = dt * rocket_eq(y[i-1] + 0.5*k1, t[i-1] + 0.5*dt, Cd, A, rho, g, mass_function)
        k3 = dt * rocket_eq(y[i-1] + 0.5*k2, t[i-1] + 0.5*dt, Cd, A, rho, g, mass_function)
        k4 = dt * rocket_eq(y[i-1] + k3, t[i-1] + dt, Cd, A, rho, g, mass_function)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
        s[i] = s[i-1] + dt * path_eq(y[i-1], t[i-1], Cd, A, rho, g)
    return s, y[:, 0]

# Funkcja zmiany masy w czasie
def mass_function(t):
    if 2800 - 57 *t >=0:
        return 5900-(2800-57*t)
    else:
        return 3100

# Parametry rakiety
Cd = 0.5
A = 1.0 * math.pi
rho = 1.2
g = 9.81

# Warunki pocztkowe
initial_velocity = 400
initial_angle = np.radians(89)
initial_mass = 5900.0
initial_conditions = np.array([initial_velocity, initial_angle, initial_mass])

# Czas symulacji
t = np.linspace(0, 300, 3000)

# Rozwizanie rwnania ruchu numerycznie - Metoda Eulera
solution_euler, velocity_euler = euler_method(
    rocket_equation, path_equation, initial_conditions, t, Cd, A, rho, g, mass_function
)

# Rozwizanie rwnania ruchu numerycznie - Metoda Rungego-Kutty (RK4)
solution_rk4, velocity_rk4 = runge_kutta_method(
    rocket_equation, path_equation, initial_conditions, t, Cd, A, rho, g, mass_function
)

# Wykres prdkoci w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, velocity_euler, label='Euler')
plt.plot(t, velocity_rk4, label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prdko (m/s)')
plt.title('Ruch rakiety balistycznej - Prdko w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres drogi w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler, label='Euler')
plt.plot(t, solution_rk4, label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Dystans (m)')
plt.title('Ruch rakiety balistycznej - Droga w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# wersja 3

import numpy as np
import matplotlib.pyplot as plt

# Parametry rakiety
m0 = 5900  # pocztkowa masa rakiety [kg]
mdot = 57  # tempo spalania paliwa [kg/s]
g = 9.81  # przyspieszenie ziemskie [m/s^2]
Thrust = 160000  # zwikszony cig silnika [N]
burnout_mass = 2800  # masa rakiety po wyczerpaniu paliwa [kg]
rho = 1.2  # Gsto powietrza [kg/m^3]
Cd = 1.1  # Zmniejszony wspczynnik oporu aerodynamicznego
A = 3.5  # Powierzchnia przekroju poprzecznego rakiety [m^2]

# Warunki pocztkowe
v0 = 10  # pocztkowa prdko [m/s]
y0 = 0  # pocztkowa wysoko [m]
theta0 = 90  # pocztkowy kt [stopnie] (rakieta zwrcona do gry)
t0 = 0  # pocztkowy czas [s]


# Funkcje opisujce rwnania ruchu
def mass(t):
    return m0 - mdot * t if m0 - mdot * t >= burnout_mass else burnout_mass


def rocket_eq(t, state):
    m = mass(t)
    # Zmiana kta po 20 sekundach
    if t >= 25:
        dthetadt = np.radians(-1.2)  # Prdko zmiany kta
    else:
        dthetadt = 0  # Kt nie zmienia si

    dvdt = ((-m * g - 0.5 * rho * state[0] * np.abs(state[0]) * Cd * A + Thrust) / m) if m > burnout_mass else -g
    dydt = state[0]

    return np.array([dvdt, dydt, dthetadt])


# Metoda Eulera do rozwizania rwna rniczkowych
def euler_method(h, T):
    num_steps = int(T / h)
    times = np.linspace(t0, T, num_steps + 1)
    states = np.zeros((num_steps + 1, 3))
    states[0, :] = [v0, y0, np.radians(theta0)]  # Zamiana stopni na radiany

    for i in range(num_steps):
        t = times[i]
        y = states[i, :]
        states[i + 1, :] = y + h * rocket_eq(t, y)

        # Sprawdzenie warunku przerwania
        if states[i + 1, 1] <= 0:
            break

    return times[:i + 2], states[:i + 2]


# Parametry symulacji
h = 0.1  # krok czasowy [s]
T = 120 # czas trwania symulacji [s]

# Rozwizanie rwna ruchu
times, states = euler_method(h, T)

# Wykresy
plt.figure(figsize=(12, 7))
plt.subplot(3, 1, 1)
plt.plot(times, states[:, 0])
plt.title('Prdko rakiety w funkcji czasu')
plt.xlabel('Czas [s]')
plt.ylabel('Prdko [m/s]')

plt.subplot(3, 1, 2)
plt.plot(times, states[:, 1])
plt.title('Wysoko rakiety w funkcji czasu')
plt.xlabel('Czas [s]')
plt.ylabel('Wysoko [m]')

plt.subplot(3, 1, 3)
plt.plot(times, np.degrees(states[:, 2]))  # Przeliczenie radianw na stopnie
plt.title('Kt rakiety w funkcji czasu')
plt.xlabel('Czas [s]')
plt.ylabel('Kt [stopnie]')

plt.tight_layout()
plt.show()
```

```python
# upoledzona wersja 4

import numpy as np, math
import matplotlib.pyplot as plt

m_loss = 57.8
fuel_mass = 2800
rocket_mass = 5900.0
max_thrust = 1000
no_fuel = fuel_mass/m_loss
S = 1.5**2*math.pi

params = {'Cd' : 0.5, 'S' : 1.5**2*math.pi, 'rho' : 1.2, 'g' : 9.81}

# Funkcja zmiany masy w czasie
def mass_function(t):
    if fuel_mass - m_loss *t > 0:
        return rocket_mass -(m_loss * t)
    else:
        return rocket_mass - fuel_mass

def fun_T(t):
    if t < no_fuel:
        return max_thrust
    else:
        return 0

# Funkcja opisujca rwnanie ruchu rakiety balistycznej
def rocket_equation(var, t, params):
    vx, vy, x, y, m = var
    g, rho, S, Cd = params['g'], params['rho'], params['S'], params['Cd']
    
    T = fun_T(t)
    theta = np.arctan2(vy,vx)
    v = np.sqrt(vx**2 + vy**2)

    dvxdt = (T * np.cos(theta) - 0.5 * rho * v**2 * Cd * S) / m
    dvydt = (T * np.sin(theta) - m * g * np.cos(theta) - 0.5 * rho * v**2 * Cd * S) / m
    dxdt = vx
    dydt = vy
    dmdt = - m_loss if t < no_fuel else 0

    return np.array([dvxdt, dvydt, dxdt, dydt, dmdt])

# Metoda Eulera
def euler_method(f, y0, t, params):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(1, n):
        y[i] = y[i-1] + dt * f(y[i-1], t[i-1], params)
    return y

# Metoda Rungego-Kutty (RK4)
def runge_kutta_method(f, y0,t, params):
    n = len(t)
    dt = t[1] - t[0]
    y = np.zeros((n, len(y0)))
    y[0] = y0
    for i in range(1, n):
        k1 = dt * f(y[i-1], t[i-1], params)
        k2 = dt * f(y[i-1] + 0.5*k1, t[i-1] + 0.5*dt, params)
        k3 = dt * f(y[i-1] + 0.5*k2, t[i-1] + 0.5*dt, params)
        k4 = dt * f(y[i-1] + k3, t[i-1] + dt, params)
        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6
    return y

# Warunki pocztkowe
theta = 45
initial_velocity = 450
vx0 = initial_velocity * np.cos(np.radians(theta))
vy0 = initial_velocity * np.sin(np.radians(theta))
x0, y0 = 0, 0
initial_conditions = np.array([vx0, vy0, x0, y0, rocket_mass])

# Czas symulacji
t_k = 85
t = np.linspace(0, t_k, int(t_k / 0.5 + 1))


# Rozwizanie rwnania ruchu numerycznie - Metoda Eulera
solution_euler = euler_method(
    rocket_equation, initial_conditions, t, params)

# Rozwizanie rwnania ruchu numerycznie - Metoda Rungego-Kutty (RK4)
solution_rk4 = runge_kutta_method(
    rocket_equation, initial_conditions, t, params)

# Wykres przemieszczenia
plt.figure(figsize=(10, 6))

plt.plot(solution_euler[:,2], solution_euler[:,3], label='Euler')
plt.plot(solution_rk4[:,2], solution_rk4[:,3], label='RK4')
plt.xlabel('Dystans')
plt.ylabel('Wysoko')
plt.title('Ruch rakiety balistycznej - przemieszczenie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres prdkoci x w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,0], label='Euler')
plt.plot(t, solution_rk4[:,0], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prdko x (m/s)')
plt.title('Ruch rakiety balistycznej - Prdko x w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres prdkoci y w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,1], label='Euler')
plt.plot(t, solution_rk4[:,1], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prdko y (m/s)')
plt.title('Ruch rakiety balistycznej - Prdko y w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres prdkoci w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, np.sqrt(solution_euler[:,0]**2 + solution_euler[:,1]**2), label='Euler')
plt.plot(t, np.sqrt(solution_rk4[:,0]**2 + solution_rk4[:,1]**2), label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Prdko (m/s)')
plt.title('Ruch rakiety balistycznej - Prdko w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Wykres drogi w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,2], label='Euler')
plt.plot(t, solution_rk4[:,2], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Dystans x (m)')
plt.title('Ruch rakiety balistycznej - Droga w czasie')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# wykres wysokoci w czasie
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,3], label='Euler')
plt.plot(t, solution_rk4[:,3], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('Wysoko y (m)')
plt.title('Ruch rakiety balistycznej - Wysoko w czasie')
plt.legend()
plt.grid(True)
plt.show()

# wykres masy
plt.figure(figsize=(10, 6))

plt.plot(t, solution_euler[:,4], label='Euler')
plt.plot(t, solution_rk4[:,4], label='RK4')
plt.xlabel('Czas (s)')
plt.ylabel('masa')
plt.title('ubytek masy')
plt.legend()
plt.grid(True)
plt.show()
```

W  modelu przedstawionym w artykule [[1](#bibliografia)] traktujemy rakiet jako punkt materialny o 3 stopniach swobody (ignorujemy stopnie swobody zwizane z obrotai w przestzreni). Wtedy przyspieszenie mona przedstawi jako sum:
$$
    \ddot{r} = \dot{v} =  a_T + a_C + a_D + a_A 
$$,
gdzie : 

$
r -\text{wektor pooenia rakiety balistycznej} \\
v - \text{wektor prdkoci rakiety balistycznej} \\
a_T - \text{wektor przyspieszenia zwizany z cigiem silnika rakietowego} \\
a_C - \text{wektor przyspieszenia sterujcego} \\
a_D - \text{wektor przyspieszenia zwizany z oporem atmosfery} \\
a_G - \text{wektor przyspieszenia grawitacyjnego} \\
a_A - \text{wektor przyspieszenia pozornego (odrodkowego i Coriolisa)} 
$ 

Przyspieszenie zwizane z oporem powietrza (poniej 100 km od powierzchni Ziemi) mona opisa wzorem:

$$ a_D = - \frac{\rho_0 e^{-kh}c_D S}{2(m_0 - \int_0^{t_b} \dot{m_f} \operatorname{d}\!t)g}||v||v
$$,
gdzie:

$\rho_0 = 1,22 kg/m^3 - \text{gsto atmosfery na wysokoci 0 m n.p.m.,} \\
k = 0,14141\cdot 10^{-3} m^{-1} \text{ - wspczynnik} \\
h - \text{wysoko lotu rakiety} \\
c_D - \text{wspczynik oporu aerodynamicznego kaduba} \\
S - \text{powierzchnia charakterystyczna} \\
m_0 - \text{masa stratowa rakiety} \\
\dot{m_f} - \text{wydatek masy paliwa} \\
t_b - \text{ czas pracy silnika rakietowego} \\
g - \text{przyspieszenie ziemskie}
$ 

Przyspieszenie grawitacyjne rakiety przedstawiono jako:

$$ a_G = - \frac{GMr}{||r||^3} $$

przyspieszenie pozorne :

$$ a_A = -2(\omega \times v)-\omega (\omega \times r) $$

przyspieszenie od siy cigu:

$$ a_T = \frac{T}{(m_0 - \int_0^{t_b} \dot{m_f} \operatorname{d}\!t)} \frac{v}{||v||} $$

gdzie: $ T $ - cig silnika rakietowego,

a wektor przyspieszenia sterujcego opisano jako:

$$ a_C = \frac{aT}{2}\left[1+\cos\left(\frac{\pi t}{t_b - t_v}\right)\right]\delta $$

gdzie:

$\delta$ - jednostkowy wektor sterujcy,

$a$ - wspczynnik korygujcy,

$t_b$ - czas pracy silnika rakietowego,

$t_v$ - czas lotu pionowego,

co zapewnia pynn regulacj kta pochylenia $\gamma$ wektora prdkoci w chwili wyczenia napdu, co pokazano na rysunku 2 [[1](#bibliografia)].
