```python
import math

import numpy as np
import pylab as py

def RK4(f, x0, t0, tk, h):
    """Metoda numeryczna Runge-Kutta czwartego rzędu"""
    t = np.arange(t0, tk, h)
    N = len(t)
    if hasattr(x0, "__len__"):
        x = np.zeros((N, len(x0)))
    else:
        x = np.zeros(N)
    x[0] = np.array(x0)
    for i in range(1, N):
        k1 = h * f(t[i-1], x[i-1])
        k2 = h * f(t[i-1] + h*0.5, x[i-1] + 0.5*k1)
        k3 = h * f(t[i-1] + h*0.5, x[i-1] + 0.5*k2)
        k4 = h * f(t[i-1] + h, x[i-1] + k3)
        x[i] = np.array(x[i-1] + (k1 + 2.0*k2 + 2.0*k3 + k4) / 6)
    return t, x

def oscylator(t, y):
    f0 = 1.0
    w0 = 1.0
    Gamma = 1.0
    W = oscylator.W

    x = y[0]
    xdot = y[1]
    return np.array([xdot, f0 * math.cos(W * t) - Gamma * xdot - w0**2 * x])

oscylator.W = 1.0

# Użyj argumentu color, aby ustawić kolor na żółty
py.plot(*RK4(oscylator, [0.0, 1.0], 0.0, 20.0, 0.1))
py.title('Metoda RK4')
py.show()

#metoda Explicit Euler
#f - funkcja f z rownania
#x0-wartosc poczatkowa
#t0-czas poczatkowy
#tk-czas koncowy
#h-krok czasowy

def EE(f,x0,t0,tk,h):
    #generujemy wektor czasow
    t=np.arange(t0,tk,h)

    #liczba krokow czasowych
    N=len(t)

    #wektor wynikowy
    if hasattr(x0, "__len__"): x=np.zeros((N,len(x0))) # gdy mamy do czynienia w równaniem wektorowym
    else: x=np.zeros(N) #dla przypadku skalarnego

    #wpisujemy wartosc poczatkowa
    x[0]=np.array(x0)
    #index
    i=1
    #petla glowna
    while (i<N):
        x[i]=np.array(x[i-1]+h*f(t[i-1],x[i-1]))
        i+=1
    return t,x

oscylator.W = 1.0
t,x=EE(oscylator,[1.0,1.0],0.0,20,0.01)
py.plot(t,x[:,0])
py.title('Metoda Eulera')
py.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def funkcja_pochodna(t, y):
    # Definicja funkcji różniczkowej dy/dt
    return -0.5 * y

def metoda_eulera(f, y0, t0, tn, h):
    # Implementacja metody Eulera
    # f - funkcja różniczkowa
    # y0 - warunek początkowy
    # t0 - początkowy czas
    # tn - końcowy czas
    # h - krok czasowy

    # Inicjalizacja list do przechowywania wyników
    t_values = np.arange(t0, tn + h, h)
    y_values = [y0]

    # Iteracyjne obliczanie wartości przy użyciu metody Eulera
    for t in t_values[:-1]:
        y_next = y_values[-1] + h * f(t, y_values[-1])
        y_values.append(y_next)

    return t_values, y_values

# Parametry
t0 = 0          # Początkowy czas
tn = 5          # Końcowy czas
y0 = 1          # Warunek początkowy
h = 0.1         # Krok czasowy

# Wywołanie metody Eulera
t_values, y_values = metoda_eulera(funkcja_pochodna, y0, t0, tn, h)

# Wykres wyników
plt.plot(t_values, y_values, label='Metoda Eulera')
plt.xlabel('Czas')
plt.ylabel('Wartość y')
plt.legend()
plt.show()
```

```python
import numpy as np, math
import matplotlib.pyplot as plt

def funkcja_pochodna(t, y, v):
    # Definicja funkcji różniczkowej dy/dt
    # v to zmienna pomocnicza, czyli dy/dt
    return v, y**2  # Przykładowa funkcja dla równania y'' = ...

def metoda_eulera(f, y0, v0, t0, tn, h):
    # Implementacja metody Eulera dla układu równań pierwszego rzędu
    # f - funkcja różniczkowa
    # y0, v0 - warunki początkowe dla y i v
    # t0 - początkowy czas
    # tn - końcowy czas
    # h - krok czasowy

    t_values = np.arange(t0, tn + h, h)
    y_values = [y0]
    v_values = [v0]

    for t in t_values[:-1]:
        y_next, v_next = y_values[-1], v_values[-1]
        dy, dv = f(t, y_next, v_next)
        y_values.append(y_next + h * dy)
        v_values.append(v_next + h * dv)

    return t_values, y_values

# Parametry
t0 = 1
tn = 100
y0 = -1
v0 = 1 # Początkowa prędkość, czyli dy/dt
h = 1

# Wywołanie metody Eulera dla układu równań
t_values, y_values = metoda_eulera(funkcja_pochodna, y0, v0, t0, tn, h)

# Wykres wyników
plt.plot(t_values, y_values, label='Metoda Eulera')
plt.xlabel('Czas')
plt.ylabel('Wartość y')
plt.legend()
plt.show()
```

$$

$
r_y \text{ - współrzędna pionowa położenia} \\
v_y \text{ - współrzędna pionowa prędkości} \\

v_x \text{ - współrzędna pozioma prędkości} \\
\theta \text{ - kąt odchylenia rakiety od pionu} \\
m \text{ - masa rakiety (zależna od czasu)} \\

S \text{ - powierzchnia przekroju poprzecznego rakiety} \\
\rho_0 = 1.22 \frac{kg}{m^3} \text{ - stała gęstości powietrza (wartość dla poziomu morza)} \\
C_d \text{ - współczynnik oporu powietrza (my przyjęliśmy 0.5)} \\
T \text{ - siła ciągu silnika rakietowego} \\
